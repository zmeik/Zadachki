# Задачка1

### В деревне живёт 30 правдиков и 30 вруликов (и больше никого). Правдики говорят только правду, а врулики - только ложь.
### У каждого  правдика есть хотя бы один друг, а у каждого врулика ровно один друг. Однажды ровно половина жителей
### деревни сказала: "Среди моих друзей нет ни одного врулика!" А все остальные жители ответили: "Среди моих друзей
### нет ни одного правдика!" Какое наименьшее количество вруликов может дружить с правдиками?

## Решение:

### Обозначим количество Вруликов которые дружат с Правдиками за "(x)". Оставшиеся Врулики, дружат с Вруликами. Так как в условии написано, что "у каждого врулика ровно один друг", а всего "В деревне живёт 30 правдиков и 30 вруликов (и больше никого)." Можно сделать и в том числе вывод, что оставшаяся часть Вруликов дружит парами 1:1 с Вруликами и только. Обозначим эту оставшуюся часть за "(y)". При чём, общее количество Вруликов выражается суммой (x + y = 30). ПЕРВОЕ УРАВНЕНИЕ. 
### Так как всего Вруликов в деревне 30 и "у каждого врулика ровно один друг". Количество Правдиков, которые дружат с Вруликами соответствено не может превышать "(х)", но и не может быть меньше (0), т.е. если обозначить за "(z)", (z <= x and z >= 0). ВТОРОЕ УРАВНЕНИЕ.
### Так как "у каждого врулика ровно один друг". Оставшейся части Правдиков остаётся дружить друг с другом. Таких Правдиков в деревне "(30 - z)". И они говорят, что "Среди моих друзей нет ни одного врулика!", потому что "Правдики говорят только правду". Но не только они могут такое сказать. Все Врулики говорят "только ложь", а это значит что те Врулики, которые дружат с Вруликами говорят тоже самое "Среди моих друзей нет ни одного врулика!", хотя и лгут. И больше никто не может этого сказать, так как Врулики которые дружат с Правдиками, говорят другую ложь: "Среди моих друзей нет ни одного правдика!". Соответственно, так как в условии задачи написанно,  что "Однажды ровно половина жителей деревни сказала: "Среди моих друзей нет ни одного врулика!"" получается, что (30 - z + у = 30), т.е. (y = z) ТРЕТЬЕ УРАВНЕНИЕ.

### Система уравнений из двух неравенств и двух выражений, с тремя неизвестными:

### x + y = 30
### z <= x and z >= 0
### y = z

### Можно заметить так же, что так как "Правдики говорят только правду" Правдик у которого есть хотя бы один друг Врулик, не может сказать: "Среди моих друзей нет ни одного врулика!". (Это значит что такой Правдик подружившись с Вруликом лишается возможности дружить с Правдиками.) Соответственно такие Правдики должны сказать: "Среди моих друзей нет ни одного правдика!" Но это можно уже не рассматривать так как три уравнения для трёх неизвестных, хватает что бы установить наименьшее "(х)". Для чего упростим систему подставив "(z = у)", и далее (у = 30 - х) построим график

### y = 30 - x
### y <= x and y >= 0

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,30,100)
fig, ax = plt.subplots()
plt.rcParams['lines.linewidth'] = 5
y1 = 30 - x
plt.plot(x, y1, '-r', label='y=30-x', zorder=3, lw=2)
#plt.title('Graph of y=30 - x')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')

plt.rcParams['lines.linewidth'] = 3
y2 = x
y3 = 0 + (x*0)

line, = ax.plot(x, y1)
#plt.plot(x, y1, '-g', label='y<=x')
plt.plot(x, y2, '-g', label='y<=x', zorder=3, lw=3)
plt.plot(x, y3, '-g', label='y>=0')
plt.axvline(x=30, zorder=1)
line.set_drawstyle("steps-pre")

plt.fill_between(x, y2, y3, where=y2>y3, color='grey', alpha=0.5)
plt.title('Graph')
plt.xlabel('x - Врулики с Правдиками', color='Blue')
plt.ylabel('y - Врулики с Вруликами', color='Blue')
plt.legend(loc='upper left')
plt.grid(which='major',linestyle='--',linewidth='0.3',color='red')
plt.minorticks_on()
plt.grid(which='minor',linestyle='-',alpha=0.2)

xi = 30/2
yi = 30/2 #Calculated values of interseption
plt.scatter(xi,yi, color='Blue', s=300, zorder=10)
plt.show()
```
